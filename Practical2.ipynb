```python
z = np.abs(stats.zscore(df['Reading_Score']))
```

```python
print(z)
```

```python
threshold = 0.18
```

```python
sample_outliers = np.where(z <threshold)
```

```python
sample_outliers
```

```python
sorted_rscore= sorted(df['Reading_Score'])
```

```python
sorted_rscore
```

```python
q1 = np.percentile(sorted_rscore, 25)
q3 = np.percentile(sorted_rscore, 75)
print(q1,q3)
```

```python
IQR = q3-q1
```

```python
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
new_df=df
for i in sample_outliers:new_df.drop(i,inplace=True)
new_df
```
